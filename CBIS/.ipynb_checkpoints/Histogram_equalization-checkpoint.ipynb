```python
import numpy as np
import cv2 as cv
import pandas
import json
import glob, os
import matplotlib.pyplot as plt
```

```python
os.chdir("F:/Dataset/Mammography/3_4")
for file in glob.glob("*.png"):
    # x = file.split(".")
    # print(x[0])
    if "mask" in file:
        image = cv.imread(f'F:/Dataset/Mammography/3_4/{file}')
        image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
        im_bw = cv.threshold(image, 1, 255, cv.THRESH_BINARY)[1]
        norm = np.zeros((image.shape[0],image.shape[1]))
        norm_image = cv.normalize(im_bw,norm,0,1,cv.NORM_MINMAX)
#         imagem_binaria=image
#         for i in range(image.shape[0]):
#             for j in range(image.shape[1]):
#                 if(image[i][j] !=0):
#                     imagem_binaria[i][j]=1
#                 else:imagem_binaria[i][j]=0
        # clahe = cv.createCLAHE(clipLimit=2, tileGridSize=(8,8))
        # cl1 = clahe.apply(image)
        print(file)
        # cv.imwrite(f'F:/Dataset/Mammography/3_5/{file}',norm_image)
        cv2.imwrite(f'F:/Dataset/Mammography/3_5/{file}', norm_image, [cv2.IMWRITE_PNG_BILEVEL, 1])
#     else:
#         image = cv.imread(f'F:/Dataset/Mammography/3_4/{file}')
#         image = cv.cvtColor(image, cv.COLOR_BGR2GRAY)

#         clahe = cv.createCLAHE(clipLimit=2, tileGridSize=(8,8))
#         cl1 = clahe.apply(image)
#         print(file)
#         cv.imwrite(f'F:/Dataset/Mammography/3_5/{file}',cl1)
#         cv.imwrite(f'F:/Dataset/Mammography/3_6/{file}',cl1)
```
